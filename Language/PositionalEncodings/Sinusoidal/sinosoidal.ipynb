```python
import torch
import matplotlib.pyplot as plt

from encoding import SinusoidalEncodings
```

```python
PositionalEncodings = SinusoidalEncodings(d_model=128, max_len=1024)
PE = PositionalEncodings(torch.ones(1, 100, 128)).squeeze(0)
```

```python
plt.figure(figsize=(10, 4))
plt.imshow(PE[:, :128].numpy(), aspect='auto', cmap='inferno')
plt.title("Sinusoidal Positional Encodings Heatmap (100 Positions x 512 Dimensions)")
plt.xlabel("Embedding Dimension")
plt.ylabel("Position")
plt.colorbar(label="Encoding Value")
plt.tight_layout()
plt.show()
```
